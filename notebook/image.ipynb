```python
import requests
import PIL.Image as pillow
import matplotlib.pyplot as plt

from io import BytesIO
```

```python
src = 'https://opencv-python.readthedocs.io/en/latest/_images/lena.jpg'
resp = requests.get(src)

content = BytesIO(resp.content)
header = resp.headers['Content-Type'].split('/')[-1].split(';')[0].split('+')[0]
```

```python
resp.headers['Content-Type']
```

```python
img = pillow.open(content)

img
```

```python
img = plt.imread(content, header)

plt.imshow(img)
```
